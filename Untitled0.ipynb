```python
using DifferentialEquations
using JLD
using PyPlot
using OrdinaryDiffEq
using DiffEqBase
using Random, Distributions
using LinearAlgebra
using Plots
pyplot()

Random.seed!(123)

function f_motion(f) 
    fx, fy, thetta = f
    
    mu = 3 #наноНьютоны
    nu = 1 #коэффициент по времени взаимодействия клеток
    sigma_K = 0.3 #наноНьютоны
    sigma_alfa = pi/3
    K_max = mu + 3*sigma_K
    betta = 2.0
    
    K_i = abs(rand(Normal(mu, sigma_K^2)))
    thetta_i = rand(Uniform(0, 2*pi))

    mod_f_new = K_i * nu
    
    fx_new = mod_f_new * cos(thetta_i + thetta)
    fy_new = mod_f_new * sin(thetta_i + thetta)
    
    thetta_new = thetta_i + thetta
    
    f_new = [fx_new, fy_new, thetta_new]
    
end 

function f_motion_mat(matr_F) 
    
    mu = 3 #наноНьютоны
    nu = 1 #коэффициент по времени взаимодействия клеток
    sigma_K = 0.3 #наноНьютоны
    sigma_alfa = pi/3
    K_max = mu + 3*sigma_K
    betta = 2.0
    
    for i in (1:N)
        #fx, fy, thetta = f
        
        K_i = abs(rand(Normal(mu, sigma_K^2)))
        alfa = rand(Normal(0, sigma_alfa^2))
    
        thetta_i = alfa * (1 - (K_i / K_max)^betta)
        mod_f_new = K_i * nu
    
        matr_F[i] = mod_f_new * cos(thetta_i + matr_F[i+2*N])
        matr_F[i+N] = mod_f_new * sin(thetta_i + matr_F[i+2*N])
    
        matr_F[i+2*N] = thetta_i + matr_F[i+2*N]
    end
    matr_F
end 

N = 1228 #количество клеток

Lx = 200.0
Ly = 200.0
L = Lx = Ly

x_max = Lx #размеры сетки
y_max = Ly

# в расчетную область помещается ≈ 4000 клеток
N_CD4TCs = 1200 # треть из них - CD4+ T-клетки, остальные 2800 - CD8+ T-клетки
N_CTLs = 28     # 1% от 2800 (все 2800 CD8+ T-клетки не рассматриваем, а только ЦТЛ с частотой 1%)
N_TCells = N_CD4TCs + N_CTLs # всего в области будет 1228 Т-клеток (и одна дендритная клетка)

N_inf = 120 # количество зараженных клеток
Inf_cells = sample(1:N_CD4TCs, N_inf, replace = false) ;

#задаем вектор системы, в котором каждый элемент - вектор одной клетки с параметрами [x, y, andle_speed, vx, vy]
vektor_sistem_xy = [[rand(Uniform(-x_max+5, x_max-5)), rand(Uniform(-y_max+5, y_max-5)), rand(Uniform(0, 2*pi))] for i in (1:N)];



Δt = 1
t_max = 100
t = Int(div(t_max, Δt));

# Masses of bodies
# Masses of bodies
n = N

# Constant
G = 9.0
mu = 1.0

# Initial values for position and velocity
Rx=[vektor_sistem_xy[i][1] for i in (1:N)]
Ry=[vektor_sistem_xy[i][2] for i in (1:N)]
R = [Rx Ry] 

f1 = [0, 0, 0]
f = [f_motion(f1) for i in (1:N)]
Fx = [f[i][1] for i in (1:N)]
Fy = [f[i][2] for i in (1:N)]
F_angle = [f[i][3] for i in (1:N)]
F = [Fx Fy F_angle]
p = [mu, F]

function motion(du, u, p, t) 
   mu, F = p
    for i=1:size(du)[1]
        du[i, :] .= F[i, 1:2]./mu 
    end
end     
   

tspan=(0.0, 1440.0) 

prob = ODEProblem(motion, R, tspan, p);
integrator = init(prob,Euler(),dt=Δt);




# размер обрасти
LLx = 400.0
LLy = 400.0
# количество узлов
NX = 401;
NY = 401;
# параметры  уравнения
u0  = 100/60/3.14/9; # concentration
del = 0.002; # destruction
D = 166.7; # diffusion coefficient
#шаг
dx = LLx/(NX-1)
dy = LLy/(NY-1);
dt = 1;

g = D/dx^2;


# генератор источника
function source_term(gc::Array{Float64,2},dx::Float64,dy::Float64,NX::Integer,NY::Integer,Thot::Float64,xc::Float64,yc::Float64,Rc::Float64)
  for i = 1:NX
      for j = 1:NY
        if (((i*dx-xc).^2+(j*dy-yc).^2 <= Rc^2))
            gc[i,j] = Thot
        end
      end
      end
end


# алгоритм Томаса
function _tridiagonalsolve!(x, d, a, b, c, N)
    for i = 2:N
        @inbounds begin
            b[i] -= a[i-1] * c[i-1] / b[i-1]
            d[i] -= a[i-1] * d[i-1] / b[i-1]
        end
    end

    @inbounds x[N] = d[N] / b[N]

    for i = N-1:-1:1
        # Solve for x by bulk substitution
        @inbounds x[i] = (d[i] - c[i]*x[i+1]) / b[i]
    end

    return nothing
end


#ADI
u = zeros(NX,NX)
NN = NX = NY # квадратная область



a = zeros(NN-1).-g;
c = zeros(NN-1).-g;
b = zeros(NN).+(2*g+del+2/dt);
#граничные условия на неявном шаге (в случае NX=NY как для x, так и для y )
c[1] = -2*g;
a[NN-1] = -2*g
gc = zeros(NX,NX); 

anim = Animation()


Plots.contour(u, fill = true,aspect_ratio = 1);


function diff(NT)
global gc
for k=1:NT
		d = copy(gc);
		# проногка по x  
		# основная область
		for i = 1:NN
			for j = 2:NN-1
				d[i,j] += 2/dt*u[i,j] + g*(u[i,j-1]-2*u[i,j]+u[i,j+1])
			end
		end   
		#явные(по у) в направлении x граничные условия 
		for i = 1:NN
			d[i,1] +=2/dt*u[i,1]+ 2*g*(u[i,2]-u[i,1])    # сверху y = 1
			d[i,NN] +=2/dt*u[i,NN]+ 2*g*(u[i,NN-1]-u[i,NN])  # снизу y = N
		end    
		#прогонка  по x
		for j = 1:NN
			x = zeros(NX)
			_tridiagonalsolve!(x, d[:,j], a[:], b[:], c[:], NX) 
			u[:,j] = x  
		end
    
    
    
		# проногка по y  
		# свободные члены
		d = copy(gc);
		for i = 2:NN-1
			for j = 1:NN
				d[i,j] += 2/dt*u[i,j] + g*(u[i-1,j]-2*u[i,j]+u[i+1,j])
			end
		end  
   
		#явные (по x)  в направлении y граничные условия 
		for j = 1:NN
			d[1,j] +=2/dt*u[1,j]+ 2*g*(u[2,j]-u[1,j]) # слева x = 1
			d[NN,j] +=2/dt*u[NN,j]+ 2*g*(u[NN-1,j]-u[NN,j])   # справа x = N
		end
		#решение по y
		for i = 1:NN
			y = zeros(NN)
			_tridiagonalsolve!(y, d[i,:], a[:], b[:], c[:],  NX)
			u[i,:] = y
		end
		#t = div(k,60)
		#p = plot(u[1:101,50],xlims = (0, 100), ylims = (0, 1000),title = "t = $t m")  
		Plots.contour(gc, fill = true,aspect_ratio = 1)
		frame(anim)  
end
end


function animate_solver(NT)
    global gc
    for time = 1:NT
        step!(integrator)
        R = integrator.u[:]
        Rx = R[1:Int(length(R)/2)]
        Ry = R[Int(length(R)/2)+1:length(R)]
        #смена направления каждые 30 сек.
        if time%(30/Δt) == 0 
            integrator.p = [ mu, f_motion_mat(F)]
        end
        #проверка граничных условий каждую 1сек
        if time%(1/Δt) == 0 
            for i in (1:N)
                if Rx[i] >= x_max-3
                     F[i] = -abs(F[i]) #смена вектора скорости
                     F[2*N+i] = pi - F[2*N+i] #смена направления(угла)
                     integrator.p = [mu, F]
                elseif Rx[i] <= -x_max+3
                    F[i] = abs(F[i])
                    F[2*N+i] = pi - F[2*N+i]
                    integrator.p = [mu, F]
                end
                if Ry[i] >= y_max-3
                     F[i+N] = -abs(F[i+N])
                     F[2*N+i] = - F[2*N+i]
                     integrator.p = [mu, F]
                elseif Ry[i] <= -y_max+3
                    F[i+N] = abs(F[i+N])
                    F[2*N+i] =  - F[2*N+i]
                    integrator.p = [mu, F]
                end
            end
         end  
   
        if time%(60*dt)==0
            gc = zeros(NX,NX); 
            for i=1:N_inf
                xxx = Rx[Inf_cells[i]]+ Lx
                yyy = Ry[Inf_cells[i]]+ Ly
                source_term(gc,dx,dy,NX,NY,u0,xxx ,yyy,3.0)
            end
            diff(1)
        end
    end
end

for i=1:60
@time animate_solver(60)
end

#gif(anim, "diff_cell30s.gif", fps=4)




#save("data10s_1h.jld", "data", u)


#u1m =load("data1m_1h.jld")["data"];
#u30s=load("data30s_1h.jld")["data"];
#u10s=load("data10s_1h.jld")["data"];

#mean(abs.(u1m./u10s.-1))*100
#std(abs.(u1m./u10s.-1))*100




```
